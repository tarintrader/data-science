# Extracting Nested Data from a Column

**Challenge**: How many different types of genres are there? Can an app belong to more than one genre? Check what happens when you use .value_counts() on a column with nested values? See if you can work around this problem by using the .split() function and the DataFrame's [.stack() method](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.stack.html).


```python
# Number of Genres?
len(df_apps_clean.Genres.unique())
```

```python
df_apps_clean.Genres.value_counts()
```

```python
# Split the 'Genres' column and stack the resulting DataFrame
df_genres_stacked = df_apps_clean['Genres'].str.split(';', expand=True).stack()
print(f'We now have a single column with shape: {df_genres_stacked.shape}')
num_genres = df_genres_stacked.value_counts()
print(f'Number of genres: {len(num_genres)}')

# Convert the stacked Series to a DataFrame and set the index
df_genres_indexed = df_genres_stacked.reset_index(level=1, drop=True).to_frame('Genres')

# Join the new DataFrame with the original DataFrame
df_result = df_genres_indexed.join(df_apps_clean.drop('Genres', axis=1))

# Group by 'Genres' and count the number of apps in each genre
df_genre_count = df_result.groupby('Genres').size().reset_index(name='App Count')

# Set 'Genres' as the index
df_genre_count.set_index('Genres', inplace=True)

df_genres = df_genre_count.sort_values(by=['App Count'], ascending=False)[:15]

df_genres
```

```python
box_plots = px.box(df_apps_clean.loc[df_apps_clean['Type'] == "Paid"],
                   x='Category',
                   y='Price',
                   title="Paid App Pricing Strategies by Category",
)


box_plots.update_layout(
                    yaxis_title='Paid App Ballpark Revenue',
                    yaxis=dict(type='log'),
                    xaxis_tickangle=45,
                    xaxis={'categoryorder':'max descending'}
                    )




box_plots.show()
```

```python
# Filter the DataFrame for paid apps
df_paid_apps = df_apps_clean.loc[df_apps_clean['Type'] == "Paid"]

# Calculate the overall average price
overall_avg_price = df_paid_apps['Price'].mean()

# Create the box plot
box_plots = px.box(
    df_paid_apps,
    x='Category',
    y='Price',
    title="Paid App Pricing Strategies by Category",
)

# Add a horizontal line for the overall average price
box_plots.add_shape(
    type='line',
    x0=-0.5,  # X-coordinate of the line (adjusted based on the number of categories)
    x1=len(df_paid_apps['Category'].unique()) - 0.5,  # X-coordinate of the line (adjusted based on the number of categories)
    y0=overall_avg_price,  # Y-coordinate of the line
    y1=overall_avg_price,  # Y-coordinate of the line
    line=dict(color='red', width=2, dash='dash'),  # Line style
)

# Update layout
box_plots.update_layout(
    yaxis_title='Paid App Ballpark Revenue',
    yaxis=dict(type='log'),
    xaxis_tickangle=45,
    xaxis={'categoryorder': 'max descending'}
)

# Show the plot
box_plots.show()
```
